###  Makine Öğrenmesi (Machine Learning) Temel Yaklaşımları



* **Supervised Learning (Gözetimli Öğrenme):** Amacımız, girdi verileri ile çıktı (hedef) arasındaki ilişkiyi öğrenmek için algoritmalar kullanmaktır. Model, önceden etiketlenmiş verilerle eğitilir ve yeni veriler için doğru çıktıları tahmin etmeye çalışır.

* **Unsupervised Learning (Gözetimsiz Öğrenme):** Amacımız, etiketlenmemiş (çıktısı bilinmeyen) veriler içindeki gizli yapıları ve desenleri bulmaktır. Verileri birbirine benzeyen özelliklerine göre ayırmak ve **kümelemek (clustering)** en yaygın kullanım amacıdır.

---

###  Supervised Learning (Gözetimli Öğrenme) Türleri



Gözetimli öğrenme algoritmaları, tahmin etmeye çalıştıkları verinin türüne göre iki ana gruba ayrılır:

**1. Classification (Sınıflandırma)**
* **Kavram:** Kategorik (sınıflandırılabilir/ayrık) verileri birbirinden ayırma yöntemidir.
* **Nasıl Çalışır?** Verileri, modelin önceden öğrendiği kategorilere atar. Veri seti kaç farklı kategoriye sahipse, yeni gelen veriyi o kategorilerden birine yerleştirerek sınıflandırır.
* **Ne Zaman Kullanılır?** Sonuç belirli bir etiket veya grup olduğunda (Örn: E-postanın "Spam" veya "Normal" olması, bir hastanın "Hasta" veya "Sağlıklı" olması).

**2. Regression (Regresyon)**
* **Kavram:** Kategorize edilemeyen, **sürekli (continuous)** olan sayısal değerleri tahmin etme yöntemidir.
* **Nasıl Çalışır?** Değişkenler arasındaki ilişkiyi matematiksel olarak modelleyerek, gelecekteki veya bilinmeyen sayısal değerleri hesaplar.
* **Ne Zaman Kullanılır?** Sonuç sayısal bir değer olduğunda (Örn: Bir evin özelliklerine göre fiyatını tahmin etmek, sıcaklık derecesini öngörmek).

###  Verileri Ayırma: Eğitim ve Test (Train / Test Split)

Makine öğrenmesi modellerini geliştirirken, modelin ezber yapmasını önlemek ve gerçek başarısını ölçebilmek için elimizdeki **tüm veri setini (Full Dataset)** iki ana parçaya bölmemiz gerekir.

**1. Veriyi Düzenleme (Arrange Data - X ve y Kavramları)**
* **Features (Özellikler / Girdiler - X):** Modelin öğreneceği ve tahmin yaparken kullanacağı verilerdir (bağımsız değişkenler). (Görseldeki mavi sütunlar).
* **Target (Hedef / Çıktı - y):** Modelin tahmin etmeye çalıştığı asıl sonuç veya etikettir (bağımlı değişken). (Görseldeki sarı sütun).

**2. Veri Setini Bölme (Train/Test Split)**
Veriler, rastgele örnekleme (random sampling) yöntemiyle iki gruba ayrılır:
* **Eğitim Verisi (X_train & y_train):** Veri setinin büyük bir kısmıdır. Makine öğrenmesi modeline verilir ve model bu verileri kullanarak aralarındaki matematiksel ilişkiyi **öğrenir** (Use for training).
* **Test Verisi (X_test & y_test):** Veri setinin ayrılan diğer kısmıdır. Modelin daha önce hiç görmediği bu veriler, eğitilmiş modelin ne kadar doğru çalıştığını **test etmek** için kullanılır (Use for testing).

**3. Makine Öğrenmesi Sisteminin Döngüsü (ML System)**
* **Eğitim Aşaması:** Eğitim verisi (Training Data), bir makine öğrenmesi algoritmasına (ML Algorithm) verilir. Algoritma bu verilerden öğrenerek bir sistem yani **Makine Öğrenmesi Modeli (ML Model)** oluşturur.
* **Tahmin Aşaması:** Eğitilmiş bu modele yeni veriler (örneğin ayırdığımız test verisi) verildiğinde, model öğrendiği kuralları uygulayarak bir tahmin yani **Çıktı (Output)** üretir.
