### Doğrusal Regresyon (Linear Regression) Temelleri

Doğrusal regresyon, bağımlı bir değişken (hedef) ile bağımsız değişkenler (özellikler) arasında **doğrusal (çizgisel) bir ilişki** olduğunu varsayan temel bir makine öğrenmesi algoritmasıdır.


#### 1. Basit Doğrusal Regresyon (Simple Linear Regression)
Sadece bir tane girdi (X) ve bir tane çıktı (Y) olan durumdur. Amacımız, veri noktalarının arasından geçen **en iyi ve en az hatalı** düz çizgiyi çizmektir.

* **Matematiksel Modeli:** $$Y \approx \beta_0 + \beta_1X$$
* **Parametreler (Katsayılar / Coefficients):**
  * $\beta_0$: Doğrunun Y eksenini kestiği nokta (Sabit / Intercept).
  * $\beta_1$: Doğrunun eğimi (Slope). X'in Y'yi ne kadar etkilediğini gösterir.
* **Tahmin Denklemi:** Model eğitildikten sonra, yeni bir x verisi için tahmin edilen y değerini ($\hat{y}$) bulmak için şu formül kullanılır:
  $$\hat{y} = \hat{\beta}_0 + \hat{\beta}_1x$$

#### 2. Hata Nasıl Hesaplanır? (Cost / Loss Function)
Çizdiğimiz tahmin doğrusu (kırmızı çizgi) ile gerçek veri noktaları (mavi noktalar) arasındaki dikey uzaklığa **Hata (Error / Residual)** denir.

* **Artık (Residual):** $e_i = y_i - \hat{y}_i$ (Gerçek değer - Tahmin edilen değer)
* **Hata Kareler Toplamı (SSE - Sum of Squared Error):** Hataların negatif değerlerinin birbirini sıfırlamasını önlemek ve büyük hataları daha çok cezalandırmak için hataların karesi alınarak toplanır.
  $$SSE = \sum_{i=1}^{n} (y_i - \hat{y}_i)^2$$
* **Optimizasyon:** Amacımız SSE'yi (zarar fonksiyonunu) **en aza indirmektir** (minimize etmek). Bunun için fonksiyonun kısmi türevleri alınıp 0'a eşitlenerek en ideal $\beta_0$ ve $\beta_1$ katsayıları bulunur.

#### 3. Çok Değişkenli Regresyon (Multiple Linear Regression)
Gerçek hayatta bir sonucu etkileyen birden fazla faktör vardır. (Örn: Ev fiyatını tahmin ederken metrekare, oda sayısı, bina yaşı gibi birçok $X$ değişkeni olması).

* Birden fazla değişken olduğu için veriler **Matris (X)** ve **Vektör (y)** olarak ifade edilir.
* Bu durumda grafik 2 boyutlu bir çizgiden çıkarak, çok boyutlu bir **Düzlem (Plane)** halini alır. Amacımız noktalara en yakın düzlemi yerleştirmektir.


* **Katsayıları Bulmanın Formülü (Normal Equation):** Matris cebiri kullanılarak, en az hatayı veren tüm katsayıları ($\beta$ vektörünü) tek seferde bulmamızı sağlayan formül şudur:
  $$\hat{\beta} = (X^T X)^{-1} X^T y$$
  *(Burada $X^T$, $X$ matrisinin transpozesini, $^{-1}$ ise tersini ifade eder.)*
