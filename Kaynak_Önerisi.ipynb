### Makine Öğrenmesi İçin Önerilen Kaynaklar

Makine öğrenmesi algoritmalarını öğrenmek, verilerle pratik yapmak ve kodlama sürecinde destek almak için kullanabileceğin temel kaynaklar şunlardır:

**1. Temel Kütüphaneler ve Dokümantasyon**
* **Scikit-learn:** Python'daki en popüler makine öğrenmesi kütüphanesidir.
    * **Ana Sayfa:** Algoritmaların nasıl çalıştığını ve parametrelerini anlatan kapsamlı dokümantasyona sahiptir.
    * **Examples (Örnekler) Sayfası:** Çeşitli algoritmaların gerçek kod üzerinde nasıl uygulandığını ve sonuçların nasıl görselleştirildiğini adım adım gösterir.

**2. Teorik Öğrenme ve Kavramsal Temeller**
* **StatQuest with Josh Starmer (YouTube):** Karmaşık makine öğrenmesi ve istatistik kavramlarını bol görsel kullanarak, adım adım ve son derece anlaşılır bir dille (eğlenceli bir şekilde) anlatan harika bir kanaldır.
* **An Introduction to Statistical Learning (ISLR):** Makine öğrenmesi ve veri biliminin teorik/matematiksel altyapısını anlamak için dünyada en çok okunan başucu kitaplarından biridir.

**3. Veri Seti (Dataset) Bulma ve Pratik Yapma**
* **Kaggle:** Dünyanın en büyük veri bilimi topluluğudur. Projelerinde kullanmak üzere binlerce ücretsiz veri seti bulabilir, başkalarının yazdığı çözümleri (notebooks) inceleyebilir ve yarışmalara katılabilirsin.
* **İBB Açık Veri Portalı:** İstanbul Büyükşehir Belediyesi'nin sağladığı gerçek hayat verilerini içerir. Türkiye özelinde, ulaşım, çevre, demografi gibi konularda gerçek verilerle yerel projeler geliştirmek için harika bir kaynaktır.

**4. Sorun Çözme ve Proje Yönetimi**
* **Stack Overflow:** Yazdığın kodlarda aldığın hataları (error) arattığında %99 ihtimalle çözümünü bulacağın, yazılımcıların soru-cevap platformudur.
* **GitHub:** Açık kaynaklı makine öğrenmesi projelerini incelemek, yazılan kodların mantığını anlamak ve kendi projelerini bir portfolyo olarak sergilemek için kullanılan en temel platformdur.
